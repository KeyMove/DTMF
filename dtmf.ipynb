```python
def generate_sine_wave(frequency, sample_rate, duration, amplitude):
    num_samples = int(sample_rate * duration)
    time = np.linspace(0, duration, num_samples, endpoint=False)
    wave = amplitude * np.sin(2 * np.pi * frequency * time)
    return wave
def generate_cos_wave(frequency, sample_rate, duration, amplitude):
    num_samples = int(sample_rate * duration)
    time = np.linspace(0, duration, num_samples, endpoint=False)
    wave = amplitude * np.cos(2 * np.pi * frequency * time)
    return wave
```

```python
import numpy as np
import scipy.signal as signal
 
# 设计FIR滤波器的函数
def design_fir_filter(sample_rate, cutoff_freq, filter_order):
    # 计算N次幂频率采样值
    nyquist_freq = sample_rate / 2.
    # 设计滤波器的截止频率（以N次幂频率为单位）
    cutoff_freq_normalized = cutoff_freq / nyquist_freq
    # 使用scipy.signal.firwin函数设计FIR滤波器
    val= signal.firwin(numtaps=filter_order, cutoff=cutoff_freq_normalized,pass_zero=False, window='hann')
    return val
 
# 示例使用
sample_rate = 4000  # 采样率1000Hz
cutoff_freq = 700   # 截止频率300Hz
filter_order = 9  # 滤波器阶数64
a = design_fir_filter(sample_rate, cutoff_freq, filter_order)
 
# 打印滤波器系数
#print("FIR滤波器的 feedforward 系数:", b)
print("FIR滤波器的 feedback 系数:", a)
plt.plot(a)
```

```python
ser.close()
ser = serial.Serial('COM5', 460800)
rab=np.zeros(8)
rabloss=np.zeros(8)
lastmax=np.zeros(8)
lastmaxdt=np.zeros(8)
def generate_frames():
    while True:
        time.sleep(1./24)
        image = np.zeros(320*240*3,dtype=np.uint8).reshape(240,320,3)
        lasty=120
        #val=np.fft.fft(np.convolve(a,(np.frombuffer(ser.read_all(),dtype=np.uint8)-127.0).astype(np.int8)),800)[:400]
        #di=np.convolve(a,(np.frombuffer(ser.read_all(),dtype=np.uint8)-127.0).astype(np.int8))
        di=(np.frombuffer(ser.read_all(),dtype=np.uint8)-127.0).astype(np.int8)
        # lastval=1
        # for i in range(len(di)):
        #     lv=di[i]*.9
        #     di[i]=di[i]-lastval
        #     lastval=lv
        nrab=prossrab(di)
        for i in range(8):
            rabloss[i]=rabloss[i]*.995+rab[i]*0.005
            rab[i]=rab[i]*.5+nrab[i]*.5
            x=int((rab[i])/1000)
            cv2.line(image,(0,10+i*15),(x,10+i*15),(0,255,255),10)
            
            if lastmax[i]<x:
                lastmax[i]=x
                lastmaxdt[i]=48
            elif lastmaxdt[i]>0:
                lastmaxdt[i]-=1
            else:
                lastmax[i]=lastmax[i]*.95+x*.05
            cv2.line(image,(int(lastmax[i]),5+i*15),(int(lastmax[i]),15+i*15),(0,255,255),1)
            cv2.putText(image, f'{int(lastmax[i])}',(int(lastmax[i])+10,10+i*15), cv2.FONT_HERSHEY_SIMPLEX, 0.25, (75, 75, 75), 1, cv2.LINE_AA) 

            cv2.putText(image, f'{x}',(x+10,10+i*15), cv2.FONT_HERSHEY_SIMPLEX, 0.25, (255, 255, 255), 1, cv2.LINE_AA) 
        val=np.fft.fft(di[:166],512)[:256]
        max1=0
        max2=0
        m1=0
        m2=0
        for i in range(256):
            v=int(np.sqrt(np.abs(val[i]) ** 2)/50)
            if i<155 and i>40:
                if v>max1:
                    m1=i
                    max1=v
            elif i>155:
                if v>max2:
                    m2=i
                    max2=v
            cv2.line(image,(i,240-lasty),(i+1,240-v),(255,255,255),1)
            lasty=v
        cv2.line(image,(m1,0),(m1,240),(255,255,255),1)
        cv2.line(image,(m2,0),(m2,240),(255,255,255),1)
        cv2.putText(image, f'{m1}',(250,30), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (255, 255, 255), 1, cv2.LINE_AA)  
        cv2.putText(image, f'{m2}',(250,60), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (255, 255, 255), 1, cv2.LINE_AA) 
        ret, jpeg = cv2.imencode('.jpg', image)  # 将帧编码为JPEG
        if not ret:
            break
 
        yield (b'--frame\r\n'
               b'Content-Type: image/jpeg\r\n\r\n' + jpeg.tobytes() + b'\r\n')  # 生成MJPEG帧
    ser.close()

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=8000)  # 在本地主机的8000端口上运行服务器
 
```
